```python

import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# Plotting
plt.plot(x, y1, label='Sine')
plt.plot(x, y2, label='Cosine')
plt.plot(x, y3, label='Tangent')
plt.plot(x, y4, label='Exponential Decay')

# Adding the legend
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 0.12), ncol=4)

# Setting y-limits
plt.ylim(-1.5, 2)

# Show plot
plt.show()
```
